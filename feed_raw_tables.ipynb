```python
from pyspark.sql.functions import col, to_date

#Read Diagnosis csv file 
df_diagnosis=spark.read.csv("/Volumes/healthcare/default/healthcare_data/diagnosis_mapping.csv", header=True, inferSchema=True)
df_diagnosis.write.format("delta").mode("append").saveAsTable("healthcare.default.raw_diagnosis_map")
```

```python
#Read first patient csv file
path="/Volumes/healthcare/default/healthcare_data/patients_daily_file_1_2025.csv"
#path="/Volumes/healthcare/default/healthcare_data/patients_daily_file_2_2025.csv"
#path="/Volumes/healthcare/default/healthcare_data/patients_daily_file_3_2025.csv"
df_patient=spark.read.csv(path, header=True, inferSchema=True)
df_patient=df_patient.withColumn("admission_date", to_date(col("admission_date")))

display(df_patient)

df_patient.write.format("delta").mode("append").option("mergeSchema", 'true').saveAsTable("healthcare.default.raw_patients_daily")
```
